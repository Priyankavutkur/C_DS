```python
def rotate_by_one(list_of_numbers):
  j =-1
  new_numbers=[]
  for i in range(len(list_of_numbers)):
    new_numbers.append(list_of_numbers[j])
    j=j+1
  return new_numbers
```

```python
rotate_by_one([1,2,3,4,5])
```

```python
def replacing(word):
    letter_2=""
    for char in word:
        num_1=chr(ord(char)+1)
        letter_2=letter_2+num_1
        if num_1=="!":
            letter_2=letter_2.replace("!","")
    return letter_2
word= input()
replacing(word)
```

```python
def reverse(s):
    if len(s) == 0:
      return s
    else:
      return reverse(s[1:])+s[0].upper()
       
  
s = input()
reverse(s)
```
